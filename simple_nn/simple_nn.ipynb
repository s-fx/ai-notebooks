```python
import torch 
import torch.nn as nn
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, random_split
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Simple neural network
```

```python
from torchvision.datasets import MNIST
# Data augmentation
# use at least ToTensor()

augmentations = transforms.Compose([
                                    transforms.ToTensor()
                                    ])

mnist_train = MNIST(root="./", download=True, train=True, transform=augmentations)
mnist_test = MNIST(root="./", download=True, train=False, transform=augmentations)

print(f"Train Samples: {mnist_train}")
print(f"Test Samples: {mnist_test}")
```

```python
# creating validation set for model selection and hyperparameter tuning
# Generator with fixed seed to sample consistent training and validation
# samples
mnist_train, mnist_valid = random_split(mnist_train, lengths=[50000,10000], 
                                        generator=torch.Generator().manual_seed(42))

print(f"Train Samples: {len(mnist_train)}")
print(f"Validation Samples: {len(mnist_valid)}")
```

```python
# create a DataLoader
train_loader = DataLoader(mnist_train, batch_size=64, shuffle=True)
valid_loader = DataLoader(mnist_valid, batch_size=len(mnist_valid), shuffle=False)
test_loader = DataLoader(mnist_test, batch_size=len(mnist_test), shuffle=False)

# first Batch
train_features, train_labels = next(iter(train_loader))
print(f"Features Batch Shape: {train_features.size()}")
print(f"Labels Batch Shape: {train_labels.size()}")
img = train_features[33].squeeze()
label = train_labels[33]
plt.imshow(img)
plt.axis("off")
plt.show()
print(f"Label: {label}")
```

```python
# visualise Batch 
grid_first_batch = make_grid(train_features, padding=1, pad_value=0.2)
plt.figure(figsize=(15,15))
plt.imshow(grid_first_batch.permute(1,2,0))
plt.axis("off")
```

```python
# create a model
class FCNetwork(nn.Module):
    def __init__(self,input_size,class_num):
        super(FCNetwork, self).__init__()
        
        self.input = nn.Sequential(
                                    nn.Flatten(),
                                    nn.Linear(input_size,1024),
                                    nn.ReLU()
                                    )
        self.hidden = nn.Sequential(
                                    nn.Linear(1024,1024),
                                    nn.ReLU(),
                                    nn.Linear(1024,512),
                                    nn.ReLU(),
                                    nn.Linear(512,265)
                                    )
        
        self.output = nn.Linear(265,class_num)
        
    def forward(self,x):
        x = self.input(x)
        x = self.hidden(x)
        return self.output(x)
    
    # -- you can add additional custom functions if needed, e.g. 
    # calculate class predictions using the softmax function after
    # running a forward path
    def predict(self,x):
        x = self.forward(x)
        preds = nn.funcional.softma(x,dim=1)
        return preds
```

```python
model = FCNetwork(input_size=28*28, class_num=10)
model
```

```python
# define a cost function
criterion = nn.CrossEntropyLoss()
```

```python
# define optimizer
from torch.optim import SGD, Adam
optimizer = SGD(model.parameters(), lr=0.01, momentum=0.9)
```

```python
# metrics and logging
from torchmetrics.classification import Accuracy
from torch.utils.tensorboard import SummaryWriter

acc_metric = Accuracy()
logger = SummaryWriter() # not used
```

```python
epochs = 5
losses = []
losses_noisy = []

for epoch in range(epochs):
    running_loss = 0.0
    epoch_loss = [] # for plot
    model.train #train step, only setting the mode nothing else (if dropout is used f.ex)
    acc_metric.reset()
    for batch_idx, batch in enumerate(train_loader):
        samples, labels = batch
    
        optimizer.zero_grad() # zero the parameter gradients
        outputs = model(samples) # forward pass
        loss = criterion(outputs, labels) # calculate loss
        epoch_loss.append(loss)
        loss.backward() # backprop
        optimizer.step() # optimization
        
        running_loss += loss.item()
        losses_noisy.append(loss)
    
    print(f'Epoch {epoch}: loss {running_loss}')
    losses.append(sum(epoch_loss)/len(epoch_loss))

#plt.plot(np.linspace(1, epochs, epochs).astype(int), losses)
plt.plot(np.linspace(1, 3910, 3910).astype(int), losses_noisy)
```

```python
# Model on test set
# valid step
model.eval  # set model into eval mode
acc_metric.reset()  # delete information from the train step
# do not calculate gradients 
with torch.no_grad():
    for batch_idx, batch in enumerate(valid_loader):
        samples, labels = batch
        output = model(samples)
            
        # metrics
        loss = criterion(output, labels)
        acc_metric(nn.functional.softmax(output, dim=1), labels)
        print(f' Valid-Acc:{acc_metric.compute()}')
```
